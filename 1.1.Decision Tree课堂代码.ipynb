```python
#开始建模
clf = tree.DecisionTreeClassifier(criterion = "entropy")#实例化，使用交叉熵
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)#返回预测的准确度度accurancy
score
'''
集成算法：13个中随机选取几个
随机划分训练集和数据集，导入的数据不一样，决策树在建树时依靠每一个优化节点找出最优的树。之前提到过，
无论决策树模型如何进化，在分枝上本质都还是追求某个不纯度相关的指标的优化，不纯度是基于节点来计算的，也就是说，
决策树在建树时，是靠优化节点来追求一棵优化的树，但最优的节点能够保证最优的树吗？集成算法被用来解决这个问题：
sklearn表示，既然一棵树不能保证最优，那就建更多不同的树，然后从中取最好的。怎样从一组数据集中建不同的树？
每次分枝时，不使用全部特征，而是随机选取一部分特征，从中选取不纯度相关指标最优的作为分枝用的节点。
这样，每次生成的树也就不同了
'''
```

```python
feature_name = ['alcohol',
 'malic_acid',
 'ash',
 'alcalinity_of_ash',
 'magnesium',
 'total_phenols',
 'flavanoids',
 'nonflavanoid_phenols',
 'proanthocyanins',
 'color_intensity',
 'hue',
 'od280/od315_of_diluted_wines',
 'proline']

import graphviz
dot_data = tree.export_graphviz(clf
                                ,feature_names = feature_name
                                ,class_names=["琴酒","雪莉","贝尔摩德"]#逗号写在前面，方便注解
                                ,filled=True #是否填充颜色，颜色越深表示不纯度越高  ！！！需要理解决策树原理
                                ,rounded=True #框有角，框的形状
)
graph = graphviz.Source(dot_data)
graph
#不纯度为0时可以产生标签类别
```

```python
clf.feature_importances_#使用那些特征，特征的重要性
```

```python
[*zip(feature_name,clf.feature_importances_)]
#根节点对于决策树的贡献永远是最高的
```

```python
#控制随机性
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size = 0.3)#多次划分数据集和测试集
clf = tree.DecisionTreeClassifier(criterion = "entropy"
                                  ,random_state=30
                                 ,splitter='random'
                                 ,max_depth=3
                                # ,min_samples_leaf=10#用来限制叶子结点的重要性参数
                                 ,min_samples_split=10#一个节点至少包含这么多的训练样本，这个节点才允许被分枝
                                ,max_features=6 #限制分枝时考虑的特征个数，超过限制个数的特征都会舍弃，
                                                #强行设定此参数可能会导致模型学习不足
                                ,min_impurity_decrease=1#限制信息增益（父节点和子节点之间的差值）大小，信息增益小于设定数值的分枝不会发生
                                  
                                 )#实例化，使用交叉熵,控制随机性,rs可以随便选
clf = clf.fit(Xtrain, Ytrain)                                
score = clf.score(Xtest, Ytest)#返回预测的准确度度accurancy
score#一切以追求分数最高为出发点
```

```python
'''
random_state用来设置分枝中的随机模式参数，默认None，在高维度时随机性会表现更明显，低维度的数据（鸢尾花）
随机性几乎不会显现。输入任意整数，会一直长出同一棵树，让模型稳定下来。

splitter也是用来控制决策树中的随机选项的，有两种输入值，输入‘best’，决策树在分枝时虽然随机，但是还是会优先选择
更重要的特征进行分枝(重要性可以通过属性feature_importances_查看)，输入"random",决策树在分枝时会更加随机，树会更深
对训练集的拟合将会降低。这也是防止过拟合的一种方式。当你预测你的模型会过拟合，使用这两种参数来帮助你降低过拟合的可能性

在不加限制的情况下，一棵决策树会生长到衡量不纯度的最优指标，或者没有更多的特征可用为止。这种情况往往可能会导致过拟合，这就是说，它会在
训练集上表现很好，在测试集上却表现很糟糕。我们收集的样本不可能和实际状况完全一致，因此当一棵决策树对训练数据有了过于优秀的解释性，
它找出的规则也必然包含了训练时产生的噪声，并使它对位置数据的拟合不足。
为了让决策树有更好的泛化性，我们要对决策树进行剪枝。正确的剪枝策略是优化决策树算法的核心。

'''

import graphviz
dot_data = tree.export_graphviz(clf#实例化特征
                                ,feature_names = feature_name
                                ,class_names=["琴酒","雪莉","贝尔摩德"]#逗号写在前面，方便注解
                                ,filled=True #是否填充颜色，颜色越深表示不纯度越高  ！！！需要理解决策树原理
                                ,rounded=True #框有角，框的形状
)
graph = graphviz.Source(dot_data)
graph
#不纯度为0时可以产生标签类别，树变宽，因为随机性增加
```

```python
#我们的树对训练集的拟合程度如何？
score_train = clf.score(Xtrain, Ytrain)
score_train
```

```python
import matplotlib.pyplot as plt
test = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(random_state=30
                                 ,splitter='random'#上两个都是随机性相关参数
                                ,criterion = "entropy"     
                                 ,max_depth=i+1
                                # ,min_samples_leaf=10#用来限制叶子结点的重要性参数
                                # ,min_samples_split=10#一个节点至少包含这么多的训练样本，这个节点才允许被分枝
                                #,max_features=6 #限制分枝时考虑的特征个数，超过限制个数的特征都会舍弃，
                                                #强行设定此参数可能会导致模型学习不足
                                #,min_impurity_decrease=1#限制信息增益（父节点和子节点之间的差值）大小，信息增益小于设定数值的分枝不会发生
                                ##，class_weight = ,自动调整样本不平衡
                                 )#实例化，使用交叉熵,控制随机性,rs可以随便选
    clf = clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    test.append(score)
    
plt.plot(range(1,11),test,color='red',label='max_depth')
plt.legend()
plt.show()

#调参没有绝对的答案，要看数据本身
```

```python
#apply返回每个测试样本所在的叶子节点的索引
clf.apply(Xtest)
```

```python
#predict返回每个测试样本的分类,fit,score,apply,predict
clf.predict(Xtest)
```
