```python
import pandas as pd
data_ = pd.read_csv(r".\3.4.Narrativedata.csv",index_col=0)
#index_col=0将第0列作为索引，不写则认为第0列为特征

data_.head()
data_.loc[:,"Age"] = data_.loc[:,"Age"].fillna(data_.loc[:,"Age"].median())
#.fillna 在DataFrame里面直接进行填补
```

```python
data_.dropna(axis=0,inplace=True)
#.dropna(axis=0)删除所有有缺失值的行，.dropna(axis=1)删除所有有缺失值的列
#参数inplace，为True表示在原数据集上进行修改，
#为False表示生成一个复制对象，不修改原数据，默认False
# _data_ = data_.drop(axis=0,inplace=False)
```

```python
from sklearn.preprocessing import LabelEncoder
#要输入的是标签，不是特征矩阵，所以允许一维
y = data_.iloc[:,-1]                         
```

```python
le = LabelEncoder()                         #实例化
le = le.fit(y)                              #导入数据
label = le.transform(y)                     #transform接口调取结果
```

```python
label
```

```python
#属性.classes_查看标签中究竟有多少类别
le.classes_                                 
```

```python
#也可以直接fit_transform一步到位
le.fit_transform(y)                         
```

```python
#使用inverse_transform可以逆转
le.inverse_transform(label)                 
```

```python
#让标签等于我们运行出来的结果
data_.iloc[:,-1] = label
```

```python
from sklearn.preprocessing import OrdinalEncoder #用来给特征编码
 #接口categories_对应LabelEncoder的接口classes_，一模一样的功能
data_.head()
```

```python
data = data_.copy()
#取出中间这两列
OrdinalEncoder().fit(data_.iloc[:,1:-1]).categories_  
```

```python
#一步到位转换为数字编码
data_.iloc[:,1:-1] = OrdinalEncoder().fit_transform(data_.iloc[:,1:-1])
```

```python
data_.head()
```

```python
##独热编码，创建哑变量，连续的三个数字0.1.2在算法看来是连续的，所以会给算法带来误解
#S,Q,C互不相容
import pandas as pd
data
```

```python
from sklearn.preprocessing import OneHotEncoder
X = data.iloc[:,1:-1]
```

```python
#自己遍历特征看看有几类
enc = OneHotEncoder(categories='auto').fit(X)
result = enc.transform(X).toarray()
result
```

```python
#依然可以直接一步到位，但为了给大家展示模型属性，所以还是写成了三步
OneHotEncoder(categories='auto').fit_transform(X).toarray()
```

```python
#依然可以直接一步到位，但为了给大家展示模型属性，所以还是写成了三步
OneHotEncoder(categories='auto').fit_transform(X).toarray()
```

```python
#依然可以还原
pd.DataFrame(enc.inverse_transform(result))
```

```python
#返回每一个稀疏矩阵中的列
enc.get_feature_names()
```

```python
#axis=1,表示跨行进行合并，也就是将两表左右相连，如果是axis=0，就是将量表上下相连
newdata = pd.concat([data,pd.DataFrame(result)],axis=1)
```
